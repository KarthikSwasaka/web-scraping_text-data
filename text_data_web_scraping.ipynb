```python
import pandas as pd
import numpy as np
```

```python
import nltk
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# source of the data
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Tirumala_Tirupati_Devasthanams')
# parsing data/creating beautifulSoap object
soup = bs.BeautifulSoup(source,'lxml')
```

```python
# Fetching the data
# taking paragraph data from the wiki page
# extracting web page-->inspect element----><p> tag data
text  =  ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
# Data preprocessing

text = re.sub(r'\[[0-9]*\]',' ',text) # removing numbers, which are in the boxes
text = re.sub(r'\s+',' ',text) # removing additional spaces
text = text.lower() # changing the total data into lowercase
text = re.sub(r'\d',' ',text)
```

```python
text
```

```python
sentences = nltk.sent_tokenize(text) # sentence tokenization
```

```python
sentences # in the from of sentences.
```

```python
sentences = [nltk.word_tokenize(sentence) for  sentence in sentences]   # work tokenization
```
